## 關於回測

```python
!pip install backtesting
```

```python
from backtesting import Backtest, Strategy
from backtesting.lib import crossover

from backtesting.test import SMA

class SmaCross(Strategy):
    def init(self):
        price = self.data.Close
        self.ma1 = self.I(SMA, price, 10)
        self.ma2 = self.I(SMA, price, 20)

    def next(self):
        if crossover(self.ma1, self.ma2):
            self.buy()
        elif crossover(self.ma2, self.ma1):
            # self.sell()
            self.position.close()
```

```python
bt = Backtest(TSMC.tail(300), SmaCross, commission=.002, exclusive_orders=True)
stats = bt.run()
bt.plot()
```

關於 00878 定期定額報酬

```python
HDY = yf.download('00878.tw')
```

```python
HDY.head()
```

```python
HDY['Close'].tail(100).plot()
```

```python
# 假設每一次買五萬元
CASH = 50000
```

```python
# 假設定期定額的時限是2022年一整年
for i in range(12):
  print(f"2022-{i+1:02}")
```

```python
# 總共花的錢
TOTALCASH = CASH*12
```

```python
# 總共買的股數
TOTALSTOCK = 0
for i in range(12):
  d = f"2022-{i+1:02}"
  TOTALSTOCK += CASH/HDY[d]['Close'][0]
```

```python
TOTALSTOCK
```

```python
# 2022年底結算總價值/花費的金額  賠了8%
TOTALSTOCK*HDY['2022']['Close'][-1]/TOTALCASH
```

```python
HDY['2022']['Close'].plot()
```

```python
LUMPSUM = TOTALCASH/HDY['2022']['Close'][0]
```

```python
LUMPSUM*HDY['2022']['Close'][-1]/TOTALCASH
```
